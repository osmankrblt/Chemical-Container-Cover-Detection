Planned my Marker Design with this notebook

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
canvas = np.zeros((500,500))+255
plt.imshow(canvas,cmap="gray")
```

```python
c1,c2 = (21,80),(250,450) #Simule Yolov7 bounding box coordinates
```

```python
cv2.rectangle(canvas,c1,c2,(0,0,0),2) #Draw bounding box
plt.imshow(canvas,cmap="gray")
```

```python
makerHeight = c2[1] - c1[1]
makerWidth = c2[0] - c1[0]
```

```python
# draw my object maker
cv2.line(canvas,(c2[0],c1[1]),(c2[0]+30,c1[1]),(0,0,255),3)
cv2.line(canvas,(c1[0]+(makerWidth//2),int(c1[1]+makerHeight*0.2)),(c2[0],c1[1]),(0,0,255),3)
```

```python
plt.imshow(canvas,cmap="gray")
```
